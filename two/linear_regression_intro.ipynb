# 线性回归入门实验

这个 Notebook 会带你完成：

1. 构造简单数据
2. 可视化数据
3. 使用线性回归拟合直线
4. 查看模型参数
5. 进行预测
6. 理解误差

适合机器学习初学者。


```python
import numpy as np
import matplotlib.pyplot as plt

# 构造训练数据
# x: 学习时间
# y: 考试成绩

x = np.array([1,2,3,4,5,6])
y = np.array([52,55,61,68,72,79])

print("学习时间:", x)
print("考试成绩:", y)
```

```python
# 可视化数据

plt.figure(figsize=(6,4))
plt.scatter(x, y)

plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.title("Training Data")

plt.show()
```

## 线性回归核心公式

线性回归模型：

\[
y = wx + b
\]

其中：

- \(x\)：输入
- \(y\)：预测值
- \(w\)：斜率
- \(b\)：截距


```python
from sklearn.linear_model import LinearRegression

# sklearn 要求二维输入
X = x.reshape(-1,1)

# 创建模型
model = LinearRegression()

# 训练模型
model.fit(X, y)

print("训练完成")
```

```python
# 查看模型参数

w = model.coef_[0]
b = model.intercept_

print("斜率 w =", w)
print("截距 b =", b)
```

```python
# 画出拟合直线

y_pred = model.predict(X)

plt.figure(figsize=(6,4))

# 原始数据
plt.scatter(x, y, label="Real Data")

# 拟合直线
plt.plot(x, y_pred, label="Regression Line")

plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.title("Linear Regression")

plt.legend()
plt.show()
```

```python
# 预测新数据

new_hour = np.array([[7]])

pred_score = model.predict(new_hour)

print("学习7小时的预测成绩:", pred_score[0])
```

## 误差理解

真实值与预测值之间的差距：

\[
error = y - \hat y
\]

机器学习的目标：

不断调整参数 \(w,b\)，让误差越来越小。


```python
# 查看每个样本的误差

errors = y - y_pred

for i in range(len(x)):
    print(f"x={x[i]}, 真实值={y[i]}, 预测值={y_pred[i]:.2f}, 误差={errors[i]:.2f}")
```
